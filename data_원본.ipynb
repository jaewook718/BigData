```python
import pandas as pd

df1 = pd.read_csv("data1.csv")
df2 = pd.read_csv("data2.csv")

df = pd.merge(df1,df2,how = 'inner',on='ID')  #ID기준으로 합치기
df.drop(['Date_x','Date_y','ID'], axis = 1, inplace = True)
df
```

```python
df.dropna(inplace = True)  #결측치 제거
df['Incident Cost'] = df['Incident Cost'].str.replace(pat =r'[^\w]', repl=r'', regex=True ) #특수문자 제거
df
```

```python
#https://rfriend.tistory.com/521
import numpy as np
# bins = [0,30,40,50]
# bins = [0,30,50]
bins = [0,20,30,40,50]

df['Age_grouping'] = np.digitize(df['Age'], bins)
df
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['scaled_DaysLost','scaled_IncidentCost']] = scaler.fit_transform(df[['Days Lost','Incident Cost']])

df
```

```python
# df.drop(['Gender','Age','Injury Location','Days Lost','Incident Cost'], axis = 1, inplace = True)
df.drop(['Gender','Age','Days Lost','Incident Cost'], axis = 1, inplace = True)
df.sort_values('Age_grouping', ascending=True)

df
```

```python
map_Injury = {'Trunk' : 0, 'Abdomen' : 1, 'Back' : 2, 'Legs' : 3, 'Neck':4, 'Head' : 5, 'Multiple':6,
       'Eye':7, 'Arms': 8, 'Hands' : 9, 'Feet' : 10}
df['target'] = df['Injury Location'].map(map_Injury)
df
```

```python
train = df[df['Age_grouping'] <5]
test = df[df['Age_grouping'] == 5]
# train['Age_grouping'] = train['Age_grouping'].astype(object)
# test['Age_grouping'] = test['Age_grouping'].astype(object)
train.drop('Age_grouping', axis = 1, inplace = True)
test.drop('Age_grouping', axis = 1, inplace = True)
```

```python
train = pd.get_dummies(train, dummy_na = False)
train
```

```python
test= pd.get_dummies(test, dummy_na = False)
test
```
