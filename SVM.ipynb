# 파일 불러오기

```python
import pandas as pd
df = pd.read_csv('data.csv')
df
```

# 전처리

```python
# df.drop(['Gender','Age','Shift','Days Lost','Incident Cost','Injury Location'], axis = 1, inplace = True)
df.drop(['Unnamed: 0','scaled_IncidentCost','Age_grouping'], axis = 1, inplace = True)
df
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

X = df.drop('Injury Location', axis = 1).astype(float)
Y = df['Injury Location'].astype(float)
Y=np.ravel(Y)

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
X
```

# 모델링

```python
kernel= ['rbf','poly','linear','sigmoid']
gamma =  [1, 0.1, 0.01, 0.001, 0.0001]
C= [0.1, 1, 10, 100, 1000]
degree = [i for i in range(1,11)]
```

```python
poly_result = pd.DataFrame(columns=['kernel','degree','C','accuracy'])
rbf_result = pd.DataFrame(columns=['kernel','gamma','C','accuracy'])
linear_result = pd.DataFrame(columns=['kernel','gamma','C','accuracy'])
sigmoid_result = pd.DataFrame(columns=['kernel','gamma','C','accuracy'])
```

```python
from sklearn import svm, datasets
from sklearn.metrics import accuracy_score

for i in kernel :
  if i == 'poly' :
    for j in degree :
      for k in C :
        svm_clf = svm.SVC(kernel=i, degree=j, C=k).fit(X_train, y_train)
        pred = svm_clf.predict(X_test)
        accuracy = accuracy_score(y_test, pred)
        poly_result = poly_result.append(pd.DataFrame([[i,j,k,accuracy]], columns=['kernel','degree','C','accuracy']), ignore_index=True)
        
  elif i=='rbf':
    for j in gamma :
      for k in C :
        supprotVectorMachine = svm.SVC(kernel=i, gamma=j, C=k).fit(X_train, y_train)
        pred = supprotVectorMachine.predict(X_test)
        accuracy = accuracy_score(y_test, pred)
        rbf_result = rbf_result.append(pd.DataFrame([[i,j,k,accuracy]], columns=['kernel','gamma','C','accuracy']), ignore_index=True)
        
  elif i=='linear':
    for j in gamma :
      for k in C :
        supprotVectorMachine = svm.SVC(kernel=i, gamma=j, C=k).fit(X_train, y_train)
        pred = supprotVectorMachine.predict(X_test)
        accuracy = accuracy_score(y_test, pred)
        linear_result = linear_result.append(pd.DataFrame([[i,j,k,accuracy]], columns=['kernel','gamma','C','accuracy']), ignore_index=True)
  else :
    for j in gamma :
      for k in C :
        supprotVectorMachine = svm.SVC(kernel=i, gamma=j, C=k).fit(X_train, y_train)
        pred = supprotVectorMachine.predict(X_test)
        accuracy = accuracy_score(y_test, pred)
        sigmoid_result = sigmoid_result.append(pd.DataFrame([[i,j,k,accuracy]], columns=['kernel','gamma','C','accuracy']), ignore_index=True)
    
```

정확도가 가장 높은 poly커널의 svm의 혼동행렬

```python
from sklearn.metrics import f1_score

sigmoid = svm.SVC(kernel='poly', degree= 7, C = 1000).fit(X_train, y_train)
sigmoid_pred = sigmoid.predict(X_test)

sigmoid_accuracy = accuracy_score(y_test, sigmoid_pred)
sigmoid_f1 = f1_score(y_test, sigmoid_pred, average='weighted')
print('Accuracy (sigmoid Kernel): ', "%.2f" % (sigmoid_accuracy*100))
print('F1 (sigmoid Kernel): ', "%.2f" % (sigmoid_f1*100))
```

```python
from sklearn.metrics import confusion_matrix
import numpy as np

y_test = np.array(y_test)
confusion_matrix(y_test, sigmoid_pred)
```

정확도가 가장 높은 rbf커널의 svm의 혼동행렬

```python
from sklearn.metrics import f1_score

sigmoid = svm.SVC(kernel='rbf', gamma= 1, C = 1).fit(X_train, y_train)
sigmoid_pred = sigmoid.predict(X_test)

sigmoid_accuracy = accuracy_score(y_test, sigmoid_pred)
sigmoid_f1 = f1_score(y_test, sigmoid_pred, average='weighted')
print('Accuracy (sigmoid Kernel): ', "%.2f" % (sigmoid_accuracy*100))
print('F1 (sigmoid Kernel): ', "%.2f" % (sigmoid_f1*100))
```

```python
from sklearn.metrics import confusion_matrix
import numpy as np

y_test = np.array(y_test)
confusion_matrix(y_test, sigmoid_pred)
```
